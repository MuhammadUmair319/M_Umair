## <a id='toc2_'></a>[ Step-1 Import Libraries](#toc0_)

```python
import seaborn as sns
import matplotlib as plt
```

## <a id='toc3_'></a>[Step-2 Load DataSet](#toc0_)

```python
planets=sns.load_dataset("planets")
planets.head()
```

# <a id='toc4_'></a>[ Step-3 Draw a line plot](#toc0_)


```python
sns.lineplot(x="number",y="mass",data=planets).set(title='Title of plot')
```

```python
sns.lineplot(x="number",y="mass",data=planets).set(title='Title of plot')
```

## <a id='toc5_'></a>[ 2 Bar Plot](#toc0_)

```python
sns.barplot(x="number", y="orbital_period", data=planets)
```

## <a id='toc6_'></a>[3 BoxPlot](#toc0_)


```python
sns.boxplot(x="number", y="mass", data=planets)
```

## <a id='toc7_'></a>[4 Scatter Plot](#toc0_)


```python
 # make a scatterplot
sns.scatterplot(x="number", y="orbital_period", data=planets)
```

## <a id='toc8_'></a>[5 Catplot](#toc0_)


```python
 # Make a catplot
sns.catplot(x="number", y="mass", data=planets,color= "#7CFC00")
```
